```python
plt.figure(figsize=(8,5))
sns.countplot(x='Channel', data=df)
plt.title("Number of Campaigns by Marketing Channel")
plt.xticks(rotation=45)
plt.show()
```

```python
cost = df.groupby('Channel')['Cost_USD'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
cost.plot(kind='bar',color='red')
plt.title("Total Marketing Cost by Channel")
plt.ylabel("Cost (USD)")
plt.show()
```

```python
revenue = df.groupby('Channel')['Revenue_USD'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
revenue.plot(kind='bar', color='green')
plt.title("Total Revenue by Channel")
plt.ylabel("Revenue (USD)")
plt.show()
```

```python
roi = df.groupby('Channel')['ROI'].mean().sort_values(ascending=False)

plt.figure(figsize=(8,5))
roi.plot(kind='bar', color='orange')
plt.title("Average ROI by Channel")
plt.ylabel("ROI")
plt.show()
```

```python
funnel = {
    "Impressions": df["Impressions"].sum(),
    "Clicks": df["Clicks"].sum(),
    "Leads": df["Leads"].sum(),
    "Conversions": df["Conversions"].sum()
}

funnel
```

```python
plt.figure(figsize=(8,5))

plt.barh(
    list(funnel.keys()),
    list(funnel.values())
)

plt.title("Marketing Conversion Funnel")
plt.xlabel("Count")
plt.show()
```

```python
ctr = df["Clicks"].sum() / df["Impressions"].sum() * 100
lead_rate = df["Leads"].sum() / df["Clicks"].sum() * 100
conversion_rate = df["Conversions"].sum() / df["Leads"].sum() * 100

print(f"Click Through Rate (CTR): {ctr:.2f}%")
print(f"Lead Rate: {lead_rate:.2f}%")
print(f"Conversion Rate: {conversion_rate:.2f}%")
```

```python
rates = {
    "CTR": ctr,
    "Lead Rate": lead_rate,
    "Conversion Rate": conversion_rate
}

plt.figure(figsize=(7,5))
plt.bar(rates.keys(), rates.values())
plt.title("Marketing Funnel Conversion Rates")
plt.ylabel("Percentage (%)")

for i, value in enumerate(rates.values()):
    plt.text(i, value + 0.3, f"{value:.2f}%", ha="center")

plt.show()
```

```python
# Extract month and year
df['Month'] = df['StartDate'].dt.to_period('M').astype(str)

# Campaign Duration (days)
df['Campaign_Duration'] = (df['EndDate'] - df['StartDate']).dt.days
```

```python
df[['StartDate','EndDate','Month','Campaign_Duration']].head()
```

```python
monthly_cost = df.groupby('Month')['Cost_USD'].sum()

plt.figure(figsize=(12,5))
monthly_cost.plot(marker='o')
plt.title("Monthly Marketing Cost")
plt.xlabel("Month")
plt.ylabel("Cost (USD)")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
monthly_revenue = df.groupby('Month')['Revenue_USD'].sum()

plt.figure(figsize=(12,5))
monthly_revenue.plot(marker='o')
plt.title("Monthly Revenue")
plt.xlabel("Month")
plt.ylabel("Revenue (USD)")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
monthly_roi = df.groupby('Month')['ROI'].mean()

plt.figure(figsize=(12,5))
monthly_roi.plot(marker='o')
plt.title("Monthly Average ROI")
plt.xlabel("Month")
plt.ylabel("Average ROI")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    df[['Impressions','Clicks','Leads','Conversions',
        'Cost_USD','Revenue_USD','ROI']].corr(),
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Campaign_Duration'], bins=20, kde=True)

plt.title("Campaign Duration Distribution")
plt.xlabel("Days")
plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x='Cost_USD',
    y='Revenue_USD',
    hue='Channel'
)

plt.title("Cost vs Revenue by Channel")
plt.show()
```

```python
print("Total Campaigns:", len(df))
print("Total Cost: $", round(df["Cost_USD"].sum(),2))
print("Total Revenue: $", round(df["Revenue_USD"].sum(),2))
print("Average ROI:", round(df["ROI"].mean(),2))
```

```python
channel_summary = df.groupby("Channel").agg({
    "Cost_USD":"sum",
    "Revenue_USD":"sum",
    "ROI":"mean"
}).sort_values("ROI", ascending=False)

channel_summary
```
